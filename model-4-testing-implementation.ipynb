# Import Library

```python
from keras.models import load_model  # TensorFlow is required for Keras to work
from PIL import Image, ImageOps  # Install pillow instead of PIL
import numpy as np
import matplotlib.pyplot as plt
```

# Disable Scientific Notation

```python
np.set_printoptions(suppress=True)
```

# Load Model

```python
model = load_model("/kaggle/input/model-4-v3/model_4_v3.h5", compile=False)
```

# Initiate Labels

```python
class_names = ["happy", "sad"]
```

# Load Image

```python
image = Image.open("/kaggle/input/emotion-dataset-2/resized-48-48/test/happy/size_48_48_13694565514_c8b8614cb1_w_face.png").convert("L")
```

The predicted image should be **happy**

## Before convert into gray

```python
plt.imshow(image, cmap='gray')
```

## After convert into gray

```python
plt.imshow(image)
```

## Image info

```python
im_width = image.width
im_height = image.height
print("Image Width: ", im_width)
print("Image Height: ", im_height)
```

Current image size is 48 x 48

# Resizing the image

**Just in case the image is not in 48 x 48**

```python
size = (48, 48)
image = ImageOps.fit(image, size, Image.Resampling.LANCZOS)
```

# Turn image into array

```python
image_array = np.asarray(image)
```

```python
image_array.shape
```

# Normalize Image

```python
normalized_image_array = (image_array.astype(np.float32) / 127.5) - 1
```

```python
normalized_image_array.shape
```

# Array Image

```python
data = np.ndarray(shape=(1, 48, 48), dtype=np.float32)
```

```python
data.shape
```

# Load Image (Again)

```python
data[0] = normalized_image_array
```

# Prediction

```python
prediction = model.predict(data)
index = np.argmax(prediction)
class_name = class_names[index]
confidence_score = prediction[0][index]
```

```python
print("Class:", class_name[0:], end=" ")
print("Confidence Score:", confidence_score)
```
